```haskell
--Retorna la longitud de una lista
miLength :: [a] -> Int
miLength [] = 0
miLength (_:xs) = 1 + miLength xs

--miLength [1,2,3,4]


--Retorna el último elemento de una lista
miLast :: [a] -> a
miLast [a] = a
miLast (_:xs) = miLast xs

--miLast [2,5,6,7]


--Duplica cada elemento de una lista de números
duplicar :: [Int] -> [Int]
duplicar [a] = [a*2]
duplicar (x:xs) = [x*2] ++ duplicar xs

--duplicar [2,3,4,5]

--Eliminar los elementos pares de una lista
eliminarPares :: [Int] -> [Int]
eliminarPares [] = []
eliminarPares [a]
    |mod a 2 == 0 = []
    |otherwise = [a]
eliminarPares (x:xs)
    |mod x 2 == 0 = eliminarPares xs
    |otherwise = [x] ++ eliminarPares xs

--eliminarPares [2,1,5,7]


-- 2.Define estas mismas funciones pero ahora usando funciones de orden superior como map y filter

--Retorna la longitud de una lista
miLengthFn :: [a] -> Int
miLengthFn l = sum (map (\_ -> 1) l)

miLengthFn [1,3,5,7]


--Retorna el último elemento de una lista
miLastFn :: [a] -> a
miLastFn = foldr1 (\_ x  -> x)

--miLastFn [2,3,4,5]

--Duplica cada elemento de una lista de números
duplicarFn :: [Int] -> [Int]
duplicarFn l = map (*2) l

--duplicarFn [2,3,4,5]


--Eliminar los elementos pares de una lista
eliminarPares :: [Int] -> [Int]
eliminarPares l = filter odd l

--eliminarPares [2,3,4,5]






```
